### Merge the files together as one csv file

#### Read an updated df

```python
all_data = pd.read_csv("all_data.csv")
all_data.head()
```

#### Clean the data

```python
all_data.head(100)
```

```python
#identify nan and delete them
nan_df = all_data[all_data.isna().any(axis = 1)]
nan_df.head()

all_data = all_data.dropna(how = 'all') #drop  nans
all_data.head()
```

##### find or and delete

```python
all_data = all_data[all_data['Order Date'].str[0:2] != 'Or']
all_data.head()
```

##### convert data types

```python
all_data['Quantity Ordered'] = pd.to_numeric(all_data['Quantity Ordered']) #to numeric
all_data['Price Each'] = pd.to_numeric (all_data['Price Each']) #to float
```

#### Add columna

##### Add new month column

```python
all_data['Month'] = all_data['Order Date'].str[0:2]
all_data['Month'] = all_data['Month'].astype('int32')
all_data.head()
```

##### Add City Column

```python
# use the .apply()
def getcity(address):
    return address.split(',')[1] #grabs the 1st string in between the commas of Purchase address using the def fnx

def getstate(address):
    return address.split(',')[2].split(' ')[1]#grabs the state as well

all_data['City'] = all_data['Purchase Address'].apply(lambda x: f"{getcity(x)} ({getstate(x)})")
all_data.head()
```

```python
all_data['Sales'] = all_data['Quantity Ordered'] * all_data['Price Each']
all_data.head()
```

### What was the best month for sale and how much was sold

```python
best_month_sale = all_data.groupby('Month').sum()
print(best_month_sale)
```

```python
#plot
months = range(1, 13)
plt.bar(months, best_month_sale['Sales'], color = 'green') #change colour
plt.xticks(months)
plt.ylabel('Sales in USD ($)')
plt.xlabel('Months')
plt.title("Sales 2019")
plt.ticklabel_format(style='plain')    # to prevent scientific notation
plt.show()

#December has the highest sales
```

#### What city has the highest number of sales

```python
highest_city_sales = all_data.groupby('City').sum()
print(highest_city_sales)
```

```python
#sort from highest to lowest
highest_city_sales_desc = highest_city_sales.sort_values(by=['Sales'], ascending=False)
print(highest_city_sales_desc)
```

```python
Cities = [city for city, df in all_data.groupby('City')]
plt.bar(Cities, highest_city_sales['Sales'], color = 'brown') #change colour
plt.xticks(Cities, rotation = 'vertical', size = 8)
plt.ylabel('Sales in USD ($)')
plt.xlabel('Months')
plt.title("Sales per City")
   
plt.show()

 # San Francisco,CA has the highest sales
```

#### What is the best time to advertise to maximize sales?

##### we can use the order date 

```python
all_data.head()
```

```python
#convert the order date into a date time object
all_data['Order Date'] = pd.to_datetime(all_data['Order Date'])
all_data.head()
```

```python
#add hour and minute columns
all_data['Hour'] = all_data['Order Date'].dt.hour
all_data['Minute'] = all_data['Order Date'].dt.minute
all_data.head()
```

```python
hours = [hour for hour, df in all_data.groupby('Hour')]

hourly = all_data.groupby('Hour').count()

plt.plot(hours, hourly, linewidth = 0.1, color = 'blue') #change colour
plt.xticks(hours)
plt.grid()
plt.xlabel('Hours')
plt.title("Sales per City")
    
plt.show()

#I recommend around 12pm and around 7pm
```
