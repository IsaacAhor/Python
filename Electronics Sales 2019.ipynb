#### Clean the data

```python
all_data.head(100) #view 100 rows of the data
```

```python
#identify nan and delete them
nan_df = all_data[all_data.isna().any(axis = 1)]
nan_df.head()

all_data = all_data.dropna(how = 'all') #drop  nans
all_data.head()
```

##### find or and delete

```python
all_data = all_data[all_data['Order Date'].str[0:2] != 'Or']
all_data.head()
```

##### convert data types

```python
all_data['Quantity Ordered'] = pd.to_numeric(all_data['Quantity Ordered']) #to numeric
all_data['Price Each'] = pd.to_numeric (all_data['Price Each']) #to float
```

#### Add columna

##### Add new month column

```python
all_data['Month'] = all_data['Order Date'].str[0:2]
all_data['Month'] = all_data['Month'].astype('int32')
all_data.head()
```

##### Add City Column

```python
# use the .apply()
def getcity(address):
    return address.split(',')[1] #grabs the 1st string in between the commas of Purchase address using the def fnx

def getstate(address):
    return address.split(',')[2].split(' ')[1]#grabs the state as well

all_data['City'] = all_data['Purchase Address'].apply(lambda x: f"{getcity(x)} ({getstate(x)})")
all_data.head()
```

```python
all_data['Sales'] = all_data['Quantity Ordered'] * all_data['Price Each']
all_data.head()
```

### What was the best month for sale and how much was sold

```python
best_month_sale = all_data.groupby('Month').sum()
print(best_month_sale)
```

```python
#plot
months = range(1, 13)
plt.bar(months, best_month_sale['Sales'], color = 'green') #change colour
plt.xticks(months)
plt.ylabel('Sales in USD ($)')
plt.xlabel('Months')
plt.title("Sales 2019")
plt.ticklabel_format(style='plain')    # to prevent scientific notation
plt.show()

#December has the highest sales
```

#### What city has the highest number of sales

```python
highest_city_sales = all_data.groupby('City').sum()
print(highest_city_sales)
```

```python
#sort from highest to lowest
highest_city_sales_desc = highest_city_sales.sort_values(by=['Sales'], ascending=False)
print(highest_city_sales_desc)
```

```python
Cities = [city for city, df in all_data.groupby('City')]
plt.bar(Cities, highest_city_sales['Sales'], color = 'brown') #change colour
plt.xticks(Cities, rotation = 'vertical', size = 8)
plt.ylabel('Sales in USD ($)')
plt.xlabel('Months')
plt.title("Sales per City")
   
plt.show()

 # San Francisco,CA has the highest sales
```

```python
#convert the order date into a date time object
all_data['Order Date'] = pd.to_datetime(all_data['Order Date'])
all_data.head()
```

#### What products are often sold together?

```python
#check the data
#pull all duplicats via Order ID

all_data.head(50)
duplicates = all_data[all_data['Order ID'].duplicated(keep = False)]
duplicates.head(50)

duplicates['Grouped'] = duplicates.groupby('Order ID')['Product'].transform(lambda x: ','.join (x))
duplicates.head()
```

```python
#drop the duplicated occurences
duplicates = duplicates[['Order ID', 'Grouped']].drop_duplicates()
duplicates.head()
```

```python
ordered_together = duplicates.groupby('Grouped').count()
ordered_together.head()
```

```python
ordered_together.sort_values(by=['Order ID'], ascending=False)
#gets the most ordered together in descending order
```

```python
#import tools
from itertools import combinations
from collections import Counter

count = Counter()
for row in duplicates['Grouped']:
    row_list = row.split(',')
    count.update(Counter(combinations(row_list, 2)))

count.most_common(10)

#another option:
#for key, value in count.most_common(10):
        #print(key, value)
    
#the most commonly sold as pair is 'iPhone', 'Lightning Charging Cable'), 1005.

#could be used for promotions and placed side by side on the shelves







```

#### What product sold the most?

```python
all_data.head()
```

```python
product_group = all_data.groupby('Product')
product_group.sum()
```

```python
quantity_ordered = product_group.sum()['Quantity Ordered']

products = [product for product, df in product_group]
plt.bar(products, quantity_ordered, color = 'indigo') #change colour
plt.xticks(products, rotation = 'vertical', size = 8)
plt.ylabel('Quantity Ordered')
plt.xlabel('Products')
plt.title("Number of Products Ordered")
   
plt.show()
```

```python
prices = all_data.groupby('Product').mean()['Price Each']
print(prices)
```

```python
fig, ax1 = plt.subplots()

ax2 = ax1.twinx()
ax1.bar(products, quantity_ordered, color = 'green')
ax2.plot(products, prices, color = 'red')

ax1.set_xlabel('Products')
ax1.set_ylabel('Quantity Ordered', color = 'green')
ax2.set_ylabel('Price($)', color = 'red')
ax1.set_xticklabels(products, rotation = 'vertical', size = 8)

plt.show()

#though macbook is the most expensive it sells better than the cheapest product, the SB charging cable
```
